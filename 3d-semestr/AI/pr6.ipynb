# 1.1.1 –ü—Ä–∏–º–µ—Ä

–ü—É—Å—Ç—å –¥–∞–Ω–∞ –Ω–∞—á–∞–ª—å–Ω–∞—è –ø–æ–ø—É–ª—è—Ü–∏—è –∏–∑ —á–µ—Ç—ã—Ä–µ—Ö —Ö—Ä–æ–º–æ—Å–æ–º —Å –¥–≤—É–º—è –≥–µ–Ω–∞–º–∏ x –∏ y. –ü–æ–∫–∞–∑–∞—Ç–µ–ª—å –∫–∞—á–µ—Å—Ç–≤–∞ —Ö—Ä–æ–º–æ—Å–æ–º—ã –æ—Ü–µ–Ω–∏–≤–∞–µ—Ç—Å—è —Ñ—É–Ω–∫—Ü–∏–µ–π Z. –ü—Ä–∏ —Ä–∞–≤–Ω–æ–º –∫–∞—á–µ—Å—Ç–≤–µ —Ö—Ä–æ–º–æ—Å–æ–º –ø—Ä–µ–¥–ø–æ—á—Ç–µ–Ω–∏–µ –æ—Ç–¥–∞–µ—Ç—Å—è —Ö—Ä–æ–º–æ—Å–æ–º–µ —Å –±–æ–ª—å—à–∏–º –Ω–æ–º–µ—Ä–æ–º. –ù–∞ –∫–∞–∂–¥–æ–º —ç—Ç–∞–ø–µ —Ö—Ä–æ–º–æ—Å–æ–º–∞ a —Å –≤—ã—Å—à–∏–º –∫–∞—á–µ—Å—Ç–≤–æ–º –ø–æ—Ä–æ–∂–¥–∞–µ—Ç —á–µ—Ç—ã—Ä–µ –Ω–æ–≤—ã—Ö —Ö—Ä–æ–º–æ—Å–æ–º—ã ùëè1,ùëê1,ùëè2,ùëê2, –æ–±–º–µ–Ω–∏–≤–∞—è—Å—å –≥–µ–Ω–∞–º–∏ —Å –¥–≤—É–º—è —Ö—Ä–æ–º–æ—Å–æ–º–∞–º–∏ b –∏ c –±–æ–ª–µ–µ –Ω–∏–∑–∫–æ–≥–æ –∫–∞—á–µ—Å—Ç–≤–∞ –ø–æ —É–∫–∞–∑–∞–Ω–Ω–æ–π —Å—Ö–µ–º–µ:    
    
–ü–æ—Å–ª–µ–¥–Ω—è—è —Ö—Ä–æ–º–æc–æ–º–∞ (—Å –Ω–∏–∑—à–∏–º –∫–∞—á–µ—Å—Ç–≤–æ–º) –≤—ã–±—ã–≤–∞–µ—Ç –∏–∑ –ø–æ–ø—É–ª—è—Ü–∏–∏. –ù–∞–π—Ç–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å –∫–∞—á–µ—Å—Ç–≤–∞ —Ö—Ä–æ–º–æ—Å–æ–º—ã –≤ –ø–æ–ø—É–ª—è—Ü–∏–∏ –∏ –æ–±—â–µ–µ –∫–∞—á–µ—Å—Ç–≤–æ –ø–æ–ø—É–ª—è—Ü–∏–∏ –ø–æ—Å–ª–µ —á–µ—Ç—ã—Ä–µ—Ö —ç—Ç–∞–ø–æ–≤ —ç–≤–æ–ª—é—Ü–∏–∏.  
–ü–æ—Ç—Ä–µ–±—É–µ—Ç—Å—è –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ñ—É–Ω–∫—Ü–∏–π –¥–ª—è —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏ –∞–ª–≥–æ—Ä–∏—Ç–º–∞. –ù–∞–ø–∏—à–µ–º –∏—Ö.  
**–†–µ—à–µ–Ω–∏–µ:**  
–ù–∞—á–Ω–µ–º —Å —Ñ—É–Ω–∫—Ü–∏–∏ –æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ —Ö—Ä–æ–º–æ—Å–æ–º—ã qZ(x,y):

```python
#—Ñ—É–Ω–∫—Ü–∏—è –∫–∞—á–µ—Å—Ç–≤–∞ —Ö—Ä–æ–º–æ—Å–æ–º
def qZ(x,y):
    return (x-3*y+1)/(3*x**2+3*y**2+1)
#—Å—É–º–º–∞ –∫–∞—á–µ—Å—Ç–≤–∞ —Ö—Ä–æ–º–æ—Å–æ–º
def qSumZ(z):
    return sum(z)
#—Å—Ö–µ–º–∞ –æ–±–º–µ–Ω–∞
def exchangeScheme(oldX,oldY,sortedId):
    X = [0 for i in range(4)]
    Y = [0 for i in range(4)]
    
    X[2] = oldX[sortedId[2]]
    X[3] = oldX[sortedId[2]]
    
    X[0] = oldX[sortedId[0]]
    
    X[1] = oldX[sortedId[1]]
    
    Y[0] = oldY[sortedId[2]]
    Y[1] = oldY[sortedId[2]]
    
    Y[2] = oldY[sortedId[0]]
    
    Y[3] = oldY[sortedId[1]]
    
    return X,Y
#—Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –º–∞—Å—Å–∏–≤–∞ –∫–∞—á–µ—Å—Ç–≤–∞
def sorting(Z):
    sortedId = sorted(range(len(Z)),key = lambda k:Z[k])
    return sortedId

def evoStep(X,Y,Z):
    _,minId = min((value,id) for (id,value) in enumerate(Z))
    X=X[:]
    Y=Y[:]
    Z=Z[:]
    
    X.pop(minId)
    Y.pop(minId)
    Z.pop(minId)
    
    return X,Y,Z

def evoSteps(X,Y,stepsNum=4):
    results=[]
    
    for i in range(4):
        arrZ = [qZ(x,Y[i]) for i,x in enumerate(X)]
        
        X,Y,Z = evoStep(X,Y,arrZ)
        X,Y = exchangeScheme(X,Y,sorting(Z))
        results.append([X,Y,qSumZ(arrZ),arrZ])
    return X,Y,results


X = [-2,-1,0,1]
Y = [-2,-1,0,1]
results = evoSteps(X,Y)
for i in range(len(results[2])):
    print(f'max_{i+1}_step: {results[2][i][2]}')
    
qualityArrZ=[]
for i in range(len(results[2])):
    qualityArrZ += results[2][i][3]

print(f'max Z:     {max(qualityArrZ)}')
```

# –ó–∞–¥–∞–Ω–∏–µ:
–í—ã–ø–æ–ª–Ω–∏—Ç–µ –ø–æ –≤–∞—Ä–∏–∞–Ω—Ç–∞–º —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—é –≥–µ–Ω–µ—Ç–∏—á–µ—Å–∫–æ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –ø—Ä–∏–ª–æ–∂–µ–Ω–Ω—ã–º–∏ –Ω–∞—á–∞–ª—å–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏.  


```python
#—Ñ—É–Ω–∫—Ü–∏—è –∫–∞—á–µ—Å—Ç–≤–∞ —Ö—Ä–æ–º–æ—Å–æ–º
def qZ(x,y):
    return (x-3*y+1)/(3*x**2+y**2+1)
#—Å—É–º–º–∞ –∫–∞—á–µ—Å—Ç–≤–∞ —Ö—Ä–æ–º–æ—Å–æ–º
def qSumZ(z):
    return sum(z)
#—Å—Ö–µ–º–∞ –æ–±–º–µ–Ω–∞
def exchangeScheme(oldX,oldY,sortedId):
    X = [0 for i in range(4)]
    Y = [0 for i in range(4)]
    
    X[2] = oldX[sortedId[2]]
    X[3] = oldX[sortedId[2]]
    
    X[0] = oldX[sortedId[0]]
    
    X[1] = oldX[sortedId[1]]
    
    Y[0] = oldY[sortedId[2]]
    Y[1] = oldY[sortedId[2]]
    
    Y[2] = oldY[sortedId[0]]
    
    Y[3] = oldY[sortedId[1]]
    
    return X,Y
#—Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –º–∞—Å—Å–∏–≤–∞ –∫–∞—á–µ—Å—Ç–≤–∞
def sorting(Z):
    sortedId = sorted(range(len(Z)),key = lambda k:Z[k])
    return sortedId

def evoStep(X,Y,Z):
    _,minId = min((value,id) for (id,value) in enumerate(Z))
    X=X[:]
    Y=Y[:]
    Z=Z[:]
    
    X.pop(minId)
    Y.pop(minId)
    Z.pop(minId)
    
    return X,Y,Z

def evoSteps(X,Y,stepsNum=4):
    results=[]
    
    for i in range(4):
        arrZ = [qZ(x,Y[i]) for i,x in enumerate(X)]
        
        X,Y,Z = evoStep(X,Y,arrZ)
        X,Y = exchangeScheme(X,Y,sorting(Z))
        results.append([X,Y,qSumZ(arrZ),arrZ])
    return X,Y,results


X = [-2,-1,0,2]
Y = [-2,0,-1,1]
results = evoSteps(X,Y)
for i in range(len(results[2])):
    print(f'max_{i+1}_step: {results[2][i][2]}')
    
qualityArrZ=[]
for i in range(len(results[2])):
    qualityArrZ += results[2][i][3]

print(f'max Z:     {max(qualityArrZ)}')
```

–ü–µ—Ä–µ—Ö–æ–¥ –Ω–∞ —Ö—É–¥—à–µ–µ (ŒîùëÜùëò=ùëÜùëò‚àíùëÜùëò‚àí1>0) —Ä–µ—à–µ–Ω–∏–µ –¥–æ–ø—É—Å—Ç–∏–º, –µ—Å–ª–∏ ùëÉ‚àó=100–≥–¥–µ —Å–Ω–∏–∂–µ–Ω–∏–µ —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –ø–æ –∑–∞–∫–æ–Ω—É ùëáùëò+1=0.5ùëáùëò –æ—Ç ùëá1=100.  
# 1.2.1 –ü—Ä–∏–º–µ—Ä
**–ó–∞–¥–∞—á–∞:**  
–ò—Ç–∞–∫, –Ω–∞—á–∞–ª—å–Ω—ã–µ —É—Å–ª–æ–≤–∏—è –∑–∞–¥–∞—á–∏ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—Ç —Å–æ–±–æ–π —Å–ª–µ–¥—É—é—â–∏–π –≥—Ä–∞—Ñ —Å —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è–º–∏ –º–µ–∂–¥—É —Ä–µ–±—Ä–∞–º–∏:  
  
**–†–µ—à–µ–Ω–∏–µ:**

```python
import networkx as nx
from math import e

distances = [(1,2,20),
            (1,3,40),
            (1,4,42),
            (1,5,33),
            (1,6,21),
            (2,3,26),
            (2,4,38),
            (2,5,42),
            (2,6,17),
            (3,4,22),
            (3,5,43),
            (3,6,21),
            (4,5,27),
            (4,6,22),
            (5,6,26)]

V = [1,4,5,2,6,3,1]
Z=[(3,4),
  (4,6),
  (5,2),
  (6,2)]
P=[49,54,43,54]
T=100

def probability(delta,T):
    return 100*e**(-delta/T)

def reductTemp(prevT):
    nexT = 0.5*prevT
    return nexT

graph = nx.Graph()
graph.add_weighted_edges_from(distances)

nx.draw_kamada_kawai(graph,node_color='#fb7258',node_size=2000,with_labels=True)
```

```python
def edgeLength(i,j,distances,roundTrip = True):
    if roundTrip:
        return max([(item[2] if (item[0]==i and item[1]==j) or (item[1]==i and item[0]==j) else -1)
                   for item in distances])
    else:
        return max([(item[2] if (item[0]==i and item[1]==j) else -1) for item in distances])
    
def routeLength(V,distances):
    edges=[]
    
    for i in range(len(V)-1):
        edges.append(edgeLength(V[i],V[i+1],distances))
    return sum(edges)

def routeOneReplacement(arrV,Z,replacementByName=True):
    decrement = 1 if replacementByName else 0
    arrV[Z[0]-decrement],arrV[Z[1]-decrement] = arrV[Z[1]-decrement],arrV[Z[0]-decrement]
    return arrV

def routeReplacement(V,Z):
    for z in Z:
        V = routeOneReplacement(V,z)
    return V

def chooseRoute(distances,V,Z,T,P):
    sumLength=routeLength(V,distances)
    arrSum=[sumLength]
    
    for i in range(len(Z)):
        newV = routeOneReplacement(V[:],Z[i])
        newS = routeLength(newV,distances)
        arrSum.append(newS)
        deltaS = newS-sumLength
        
        if deltaS>0:
            p = probability(deltaS,T)
            if p > P[i]:
                V = newV
                sumLength=newS
        else:
            V=newV
            sumLength=newS
        T = reductTemp(T)
    return V, arrSum

def drawRouteGraph(distances,bestRoute):
    newDistances = []
    
    for i in range(len(bestRoute)-1):
        for distance in distances:
            if distance[0] == bestRoute[i] and distance[1] == bestRoute[i+1] \
            or distance[1] == bestRoute[i] and distance[0] == bestRoute[i+1]:
                newDistances.append(distance)
    graph = nx.Graph()
    graph.add_weighted_edges_from(newDistances)
    nx.draw_kamada_kawai(graph,node_color='#fb7258',node_size=2000,with_labels=True)
    
bestRoute, arrLength = chooseRoute(distances,V,Z,T,P)

print(f'–õ—É—á—à–∏–π –≤—ã–±—Ä–∞–Ω–Ω—ã –º–∞—Ä—à—Ä—É—Ç: {bestRoute}')
print(f'–î–ª–∏–Ω–∞ –ª—É—á—à–µ–≥–æ –≤—ã–±—Ä–∞–Ω–Ω–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞: {routeLength(bestRoute,distances)}')
print(f'–î–ª–∏–Ω—ã –≤—Å–µ—Ö —Ä–∞—Å—Å–º–æ—Ç—Ä–µ–Ω–Ω—ã—Ö –º–∞—Ä—à—Ä—É—Ç–æ–≤: {arrLength}')

drawRouteGraph(distances,bestRoute)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from math import e

distances = [(1, 2, 24),
            (1, 3, 41),
            (1, 4, 36),
            (1, 5, 22),
            (1, 6, 19),
            (2, 3, 21),
            (2, 4, 33),
            (2, 5, 33),
            (2, 6, 14),
            (3, 4, 27),
            (3, 5, 39),
            (3, 6, 23),
            (4, 5, 20),
            (4, 6, 20),
            (5, 6, 19)] # –î–ª–∏–Ω—ã —Ä–µ–±–µ—Ä

V = [1, 3, 4, 5, 6, 2, 1] # –ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –ø—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏—è –º–∞—Ä—à—Ä—É—Ç–∞

Z = [(3,4), (4,6), (5,2), (6,2)] # –ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∑–∞–º–µ–Ω –≤–µ—Ä—à–∏–Ω

P = [33, 82, 51, 76] # –°–ª—É—á–∞–π–Ω—ã–µ —á–∏—Å–ª–∞, –≤—ã–ø–∞–≤—à–∏–µ –≤ –ø—Ä–æ—Ü–µ—Å—Å–µ —Å—á–µ—Ç–∞

T = 100 # –ù–∞—á–∞–ª—å–Ω–∞—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞

# –§—É–Ω–∫—Ü–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏
def probability(delta,T):
    return 100*e**(-delta/T)

# –§—É–Ω–∫—Ü–∏—è –∏–∑–º–µ–Ω–µ–Ω–∏—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã
def reductTemp(prevT):
    nexT = 0.5*prevT
    return nexT

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∞
G = nx.Graph()
G.add_weighted_edges_from(distances)

nx.draw_kamada_kawai(G, node_color='#fb7258', node_size=2000, with_labels=True)
```

```python
def edgeLength(i,j,distances,roundTrip = True):
    if roundTrip:
        return max([(item[2] if (item[0]==i and item[1]==j) or (item[1]==i and item[0]==j) else -1)
                   for item in distances])
    else:
        return max([(item[2] if (item[0]==i and item[1]==j) else -1) for item in distances])
    
def routeLength(V,distances):
    edges=[]
    
    for i in range(len(V)-1):
        edges.append(edgeLength(V[i],V[i+1],distances))
    return sum(edges)

def routeOneReplacement(arrV,Z,replacementByName=True):
    decrement = 1 if replacementByName else 0
    arrV[Z[0]-decrement],arrV[Z[1]-decrement] = arrV[Z[1]-decrement],arrV[Z[0]-decrement]
    return arrV

def routeReplacement(V,Z):
    for z in Z:
        V = routeOneReplacement(V,z)
    return V

def chooseRoute(distances,V,Z,T,P):
    sumLength=routeLength(V,distances)
    arrSum=[sumLength]
    
    for i in range(len(Z)):
        newV = routeOneReplacement(V[:],Z[i])
        newS = routeLength(newV,distances)
        arrSum.append(newS)
        deltaS = newS-sumLength
        
        if deltaS>0:
            p = probability(deltaS,T)
            if p > P[i]:
                V = newV
                sumLength=newS
        else:
            V=newV
            sumLength=newS
        T = reductTemp(T)
    return V, arrSum

def drawRouteGraph(distances,bestRoute):
    newDistances = []
    
    for i in range(len(bestRoute)-1):
        for distance in distances:
            if distance[0] == bestRoute[i] and distance[1] == bestRoute[i+1] \
            or distance[1] == bestRoute[i] and distance[0] == bestRoute[i+1]:
                newDistances.append(distance)
    graph = nx.Graph()
    graph.add_weighted_edges_from(newDistances)
    nx.draw_kamada_kawai(graph,node_color='#fb7258',node_size=2000,with_labels=True)
    
bestRoute, arrLength = chooseRoute(distances,V,Z,T,P)

print(f'–õ—É—á—à–∏–π –≤—ã–±—Ä–∞–Ω–Ω—ã –º–∞—Ä—à—Ä—É—Ç: {bestRoute}')
print(f'–î–ª–∏–Ω–∞ –ª—É—á—à–µ–≥–æ –≤—ã–±—Ä–∞–Ω–Ω–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞: {routeLength(bestRoute,distances)}')
print(f'–î–ª–∏–Ω—ã –≤—Å–µ—Ö —Ä–∞—Å—Å–º–æ—Ç—Ä–µ–Ω–Ω—ã—Ö –º–∞—Ä—à—Ä—É—Ç–æ–≤: {arrLength}')

drawRouteGraph(distances,bestRoute)
```

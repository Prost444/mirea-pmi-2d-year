```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)
data = np.random.rand(200)


mean_value = np.mean(data)
median_value = np.median(data)


print(f"Среднее значение: {mean_value}")
print(f"Медианное значение: {median_value}")


plt.scatter(range(len(data)), data, alpha=0.5)
plt.axhline(mean_value, color='red', linestyle='--', label='Среднее')
plt.axhline(median_value, color='green', linestyle='--', label='Медиана')
plt.legend()
plt.xlabel('Индекс элемента')
plt.ylabel('Значение')
plt.title('Точечная диаграмма рассеяния')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 100, 100)

def my_function(x):
    return np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(x**2)) / (np.abs(1 - np.sin(x)**3) + np.log(np.abs(2*x)))

y = my_function(x)

x_slice = x[:50]
y_slice = y[:50]

plt.plot(x, y, label='Функция')
plt.xlabel('x')
plt.ylabel('Значение функции')
plt.title('График функции')
plt.legend()


plt.scatter(x_slice, y_slice, label='Срез', color='red', marker='o')
plt.xlabel('x')
plt.ylabel('Значение функции')
plt.title('График среза')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps
from numpy import trapz


def my_function(x):
    return np.abs(np.cos(x * np.exp(np.cos(x) + np.log(x+1))))


x = np.arange(0, 21, 1)
y = my_function(x)


plt.plot(x, y, label='Функция')
plt.fill_between(x, 0, y, alpha=0.2)
plt.xlabel('x')
plt.ylabel('Значение функции')
plt.title('График функции с заливкой площади')
plt.legend()


area_trapz = trapz(y, x)
print(f'Площадь под кривой (метод трапеции): {area_trapz}')

plt.show()
```

```python
import tkinter as tk
import math

def calculate():
    try:
        x = float(entry_x.get())
        y = float(entry_y.get())
        operation = operation_var.get()
        
        if operation == "+":
            result.set(x + y)
        elif operation == "-":
            result.set(x - y)
        elif operation == "*":
            result.set(x * y)
        elif operation == "/":
            if y == 0:
                result.set("Ошибка: деление на ноль")
            else:
                result.set(x / y)
        elif operation == "e^(x+y)":
            result.set(math.exp(x + y))
        elif operation == "sin(x+y)":
            result.set(math.sin(x + y))
        elif operation == "cos(x+y)":
            result.set(math.cos(x + y))
        elif operation == "x^y":
            result.set(x ** y)
    except ValueError:
        result.set("Ошибка: введите корректные числа")

# Создаем главное окно приложения
root = tk.Tk()
root.title("Калькулятор")

# Создаем переменные для ввода, операции и результата
entry_x = tk.Entry(root)
entry_y = tk.Entry(root)
operation_var = tk.StringVar()
operation_var.set("+")  # Значение по умолчанию
result = tk.StringVar()

# Создаем интерфейс
label_x = tk.Label(root, text="x:")
label_y = tk.Label(root, text="y:")
label_operation = tk.Label(root, text="Операция:")
label_result = tk.Label(root, text="Результат:")

entry_x.grid(row=0, column=1)
entry_y.grid(row=1, column=1)
tk.OptionMenu(root, operation_var, "+", "-", "*", "/", "e^(x+y)", "sin(x+y)", "cos(x+y)", "x^y").grid(row=2, column=1)
label_x.grid(row=0, column=0)
label_y.grid(row=1, column=0)
label_operation.grid(row=2, column=0)
label_result.grid(row=3, column=0)

# Создаем отдельные кнопки для каждой операции
tk.Button(root, text="Сложение (+)", command=lambda: operation_var.set("+")).grid(row=4, column=0)
tk.Button(root, text="Вычитание (-)", command=lambda: operation_var.set("-")).grid(row=4, column=1)
tk.Button(root, text="Умножение (*)", command=lambda: operation_var.set("*")).grid(row=4, column=2)
tk.Button(root, text="Деление (/)", command=lambda: operation_var.set("/")).grid(row=4, column=3)
tk.Button(root, text="e^(x+y)", command=lambda: operation_var.set("e^(x+y)")).grid(row=5, column=0)
tk.Button(root, text="sin(x+y)", command=lambda: operation_var.set("sin(x+y)")).grid(row=5, column=1)
tk.Button(root, text="cos(x+y)", command=lambda: operation_var.set("cos(x+y)")).grid(row=5, column=2)
tk.Button(root, text="x^y", command=lambda: operation_var.set("x^y")).grid(row=5, column=3)

tk.Button(root, text="Вычислить", command=calculate).grid(row=6, column=1)
tk.Label(root, textvariable=result).grid(row=6, column=2)

root.mainloop()
```
